# Check for duplicate DataFrame Rows with the duplicated() Method

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df["Team"] = df["Team"].astype("category")
df.sort_values("First Name", inplace = True)
```

```python
df.head(50)

ADAM
ABRAHAM
JOSEPH
JACOB
MOSES
NOAH
GABRIEL
JESUS
DAVID
```

```python
df["First Name"]
```

```python
mask1 = df["First Name"].duplicated()  # duplicated() returns all duplicate values as True except first one
```

```python
df[mask1]
```

```python
mask2 = df["First Name"].duplicated(keep = "last")  # By Default keep = "first"
```

```python
df[mask2]
```

```python
mask3 = df["First Name"].duplicated(keep = False)
```

```python
df[mask3].head(20)
```

```python
mask4 = ~ df["First Name"].duplicated(keep = False) # '~' sign gives the opposite calculation
```

```python
df[mask4]
```

# Delete duplicate DataFrame rows with drop_duplicate() Method
