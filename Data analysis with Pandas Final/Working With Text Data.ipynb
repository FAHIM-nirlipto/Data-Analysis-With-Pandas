# The expand and n parameters of the str.split() method

```python
import pandas as pd

df = pd.read_csv('chicago.csv').dropna(how ="all")

df['Position Title'] = df['Position Title'].astype('category')

df['Department'] = df['Department'].astype('category')

df
```

```python
df['Name'].str.split(',')
```

```python
df['Name'].str.split(',', expand = True) # 'expand = True' return a DataFrame of splited values
```

```python
df[['Last Name', 'First Name']] = df['Name'].str.split(',', expand = True)
```

```python
df
```

```python
df['Position Title'].str.split(' ', expand = True)
```

```python
df['Position Title'].str.split(' ', expand = True, n = 1 )
```

```python
df[['First Title Word', 'Rest of Title']] = df['Position Title'].str.split(' ', expand = True, n = 1 )
```

```python
df
```
