# Outer Join

```python
import pandas as pd

week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customer = pd.read_csv('Restaurant - Customers.csv')
food = pd.read_csv('Restaurant - Foods.csv')
```

```python
week1
```

```python
week2
```

```python
week1.merge(week2, how = 'outer' , on = 'Customer ID', suffixes=('_week 1', '_week 2'))
```

```python
len(week1.merge(week2, how = 'outer' , on = 'Customer ID', suffixes=('_week 1', '_week 2')))
```

```python
week1.merge(week2, how = 'outer' , on = 'Customer ID', suffixes=('_week 1', '_week 2'),indicator = True)
```

```python
merged = week1.merge(week2, how = 'outer' , on = 'Customer ID', suffixes=('_week 1', '_week 2'),indicator = True)
```

```python
merged['_merge'].value_counts()
```

```python
merged['_merge'].isin(['right_only','left_only'])
```

```python
mask = merged['_merge'].isin(['right_only','left_only'])
```

```python
merged[mask]
```

# Left Join and Right Join

```python
import pandas as pd

week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customer = pd.read_csv('Restaurant - Customers.csv')
food = pd.read_csv('Restaurant - Foods.csv')
```

```python
food
```

```python
week1.merge(food, how = 'left', on = 'Food ID')
```

```python
food.merge(week1, how = 'right', on = 'Food ID')
```

# The left_on and right_on parameters

```python
import pandas as pd

week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customer = pd.read_csv('Restaurant - Customers.csv')
food = pd.read_csv('Restaurant - Foods.csv')
```

```python
week2
```

```python
customer
```

```python
week2.merge(customer, how = 'left', left_on = 'Customer ID', right_on = 'ID')
```

```python
updated = week2.merge(customer, how = 'left', left_on = 'Customer ID', right_on = 'ID').drop('ID', axis = 1)
```
