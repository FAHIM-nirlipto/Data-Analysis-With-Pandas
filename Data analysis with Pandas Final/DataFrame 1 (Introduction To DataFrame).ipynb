```python
nba.sort_values("Name")
```

```python
nba
```

```python
nba.sort_values("Name", ascending = False)
```

```python
nba.sort_values("Name", ascending = True, inplace = True)
```

```python
nba
```

```python
nba.sort_values("Salary", ascending = False)
```

```python
nba.sort_values("Salary", ascending = False, na_position = "first") # na_position = "first" returns null values from starting
```

# Sort a DataFrame with sort_values() method for multiple columns

```python
nba.sort_values(["Team","Name"], ascending = [False , True]) # ascending = [False , True] returns sorting sequence for the columns
```

# Rank Series values with the rank() method

```python
import pandas as pd

nba = pd.read_csv("nba.csv").dropna(how = "all")

nba
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
```

```python
nba
```

```python
nba["Salary"].rank()
```

```python
nba["Salary"].rank().astype("int")
```

```python
nba["Salary"].rank(ascending = False).astype("int")
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
```

```python
nba
```

```python
nba.sort_values("Salary", ascending = False,  inplace = True)
```

```python
nba
```
